```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
%matplotlib inline
```

```python
df = pd.read_csv('./dataset/business.csv')
df.head()
```

```python
df.info()
```

```python
df['date']
```

```python
## Feature Engineering
df['Date'] = df['date'].str.split('-').str[0]
df['Month'] = df['date'].str.split('-').str[1]
df['Year'] = df['date'].str.split('-').str[2]
```

```python
df.head()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop("date" , axis= 1 , inplace= True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Arrival_hour'] = df['arr_time'].str.split(':').str[0]
df['Arrival_minute'] = df['arr_time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_minute'] = df['Arrival_minute'].astype(int)
```

```python
df.info()
```

```python
df.drop("arr_time" , axis = 1 , inplace= True)
```

```python
df.head()
```

```python
df['Dep_hour'] = df['dep_time'].str.split(':').str[0]
df['Dep_minute'] = df['dep_time'].str.split(':').str[1]
```

```python
df['Dep_hour'] = df['Dep_hour'].astype(int)
df['Dep_minute'] = df['Dep_minute'].astype(int)
```

```python
df.head()
```

```python
df.drop("dep_time" , axis = 1 , inplace= True)
```

```python
df.head()
```

```python
df['Total_Stops'] = df['stop'].apply(lambda x : re.sub(r'\s+' , ' ' , x).strip().split(" ")[0])
```

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df.info()
```

```python
df.drop('stop' , axis = 1 , inplace= True)
```

```python
df.info()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop' : 0, '1-stop' : 1, '2+-stop' : 2})
```

```python
df.head()
```

```python
df['price'] = df['price'].str.replace("," , "").astype(int)
```
